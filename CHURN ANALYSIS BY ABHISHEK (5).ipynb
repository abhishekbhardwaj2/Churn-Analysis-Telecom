```python
#Tenure vs MonthlyCharges Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df_cleaned = df[(df['TotalCharges'].notnull()) & (df['MonthlyCharges'] != 0)]
sns.scatterplot(x=df_cleaned['tenure'], y=df_cleaned['MonthlyCharges'], hue=df_cleaned['Churn'])
plt.title("Tenure vs MonthlyCharges (Cleaned Data)")
plt.xlabel("Tenure")
plt.ylabel("MonthlyCharges")
plt.show()
```

```python
#Tenure vs TotalCharges Analysis
df_cleaned = df[(df['TotalCharges'].notnull()) & (df['TotalCharges'] != 0)]
sns.scatterplot(x=df_cleaned['tenure'], y=df_cleaned['TotalCharges'], hue=df_cleaned['Churn'])
plt.title("Tenure vs TotalCharges")
plt.xlabel("Tenure")
plt.ylabel("TotalCharges")
plt.show()
```
